```python
from sklearn import metrics
 
import gensim.models as g
import codecs
```

```python
model_src="../../models/enwiki_dbow/doc2vec.bin"
test_docs_src="data/input_test_docs.txt"
```

```python
#inference hyper-parameters
start_alpha=0.01
infer_epoch=1000
```

```python
#load model
m = g.Doc2Vec.load(model_src)
#read the file: put the sentences in a list: each position contains a array with a unique sentence
test_docs = [ x.strip().split() for x in codecs.open(test_docs_src, "r", "utf-8").readlines() ]
```

```python
#How many sentences?
len(test_docs)
```

```python
# X is a list containing the embeddings of all sentences
X=[]
for d in test_docs:
     
    X.append( m.infer_vector(d, alpha=start_alpha, steps=infer_epoch) )
```
