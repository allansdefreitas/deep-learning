```python
import pandas as pd
import numpy as np
```

```python
zoo = pd.read_csv('datasets/zoo.csv', delimiter=',')
zoo.head()
```

```python
zoo_eats = pd.read_csv('datasets/zoo_eats.csv', delimiter=';')
zoo_eats
```

```python
#you can do of this form too:
#zoo_eats2 = pd.DataFrame([['elephant','vegetables'], ['tiger','meat'], ['kangaroo','vegetables'], ['zebra','vegetables'], ['giraffe','vegetables']], columns=['animal', 'food'])
```

```python
# MERGE (default INNER) ---------------------------------------------------------------------------------######
#That is VERY simple! show all INTERSECTED VALUES between both tables
zoo.merge(zoo_eats)
```

```python
#ITS the same thing. Only difference is the order of the columns (food). SAME RESULT
zoo_eats.merge(zoo)
```

```python
#MERGE (OUTER) ----------------------------------------------------------------------------------------#
#Show all values (including NOT INTERSECTED values), like SQL OUTER JOIN
zoo.merge(zoo_eats, how = 'outer')
```

```python
# MERGE (LEFT) ----------------------------------------------------------------#
# INTERSECTION and LEFT values
zoo.merge(zoo_eats, how = 'outer')
```

```python
#Providing manually the columns names to association : animal-animal
zoo.merge(zoo_eats, how='left', left_on='animal', right_on='animal')
```

```python
# SORTING METHODS ------------------------------------------------------------######
```

```python
zoo.sort_values('water_need')
#or zoo.sort_values(by = ['water_need'])
```

```python
#sorting by multiple columns
zoo.sort_values(by = ['animal','water_need'])
```

```python
#DESCENDING SORT: MAX TO MIN
zoo.sort_values(by = ['water_need'], ascending=False)
```

```python
# RESET INDEX ------------------------------------------------------------------------#####
zoo.sort_values(by = ['water_need'], ascending=False).reset_index()
```

```python
#remove reseted indexes
zoo.sort_values(by = ['water_need'], ascending=False).reset_index(drop=True)
```

```python
# FILLNA --------------------------------------------------------------------------------------######
#Replace NaN with a alternative text
zoo.merge(zoo_eats, how='left').fillna('unknown')
```

```python
#cause lions eat meat
zoo.merge(zoo_eats, how='left').fillna('meat')
```
